```python
class Tasks:
    def __init__(self,description,priority):
        self.description = description
        self.priority = priority
        self.completed = False

class ToDoList:
    def __init__(self):
        self.Tasks_List = []
        
    def add_tasks(self,task):
        self.Tasks_List.append(task)
        
    def remove_tasks(self,task):
        self.Tasks_List.remove(task)
        
    def mark_task_completed(self,task):
        task.completed = True
        
    def print_tasks(self):
        print("To-Do-List : ")
        for i in self.Tasks_List:
            status = "Completed." if i.completed else "Not Completed."
            print(f"Description : {i.description} , Priority : {i.priority} , Status : {status}")
                
            
            
if __name__ == "__main__":
    to_do_list = ToDoList()    #object of ToDOList()
    
    while True:
        print("\n1. Add Task")
        print("2. Print Tasks")
        print("3. Mark Task as Completed")
        print("4. Exit \n")
        
        choice = int(input("Enter Your Choice (1-4) : "))
        
        if(choice == 1):
            description = input("Enter Task Description : ")
            priority = input("Enter Task Priority : ")
            new_task = Tasks(description,priority)
            to_do_list.add_tasks(new_task)
            print("Task Added Successfully..! \n")
            
            
        elif(choice == 2):
            to_do_list.print_tasks()
            
            
        elif(choice == 3):
            to_do_list.print_tasks()
            task_index = int(input("Enter the index of the task to mark as completed : "))
            if 0 <= task_index < len(to_do_list.Tasks_List):
                to_do_list.mark_task_completed(to_do_list.Tasks_List[task_index])
            else :
                print("Invalid index please try again. \n")
              
            
        elif(choice == 4):
            print("Exiting Program.\n")
            break
        
        else :
            print("Invalid choice please enter a choice between 1 and 4")
```
